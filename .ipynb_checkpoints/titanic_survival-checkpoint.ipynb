```python
# import libraries
# python staples
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
# scikitlearn
from sklearn import linear_model
import seaborn as sns
# use matplot in jupyter notebook
%matplotlib inline
```

```python
# grab the data and put it in a pandas dataframe
df_train_all = pd.read_csv('train.csv')
df_test_all = pd.read_csv('train.csv')
# print some info
print( 'example:\n', df_train_all.iloc[0])
# grab survival as desired
y_train = df_train_all.Survived
y_test = df_test_all.Survived
# some data is pointless, get rid of id
df_x_train = df_train_all.drop(
    ['PassengerId','Name','Ticket', 'Cabin'], axis=1 )
df_x_test = df_test_all.drop(
    ['PassengerId','Name','Ticket', 'Cabin'], axis=1 )
```

```python
# clean it
# find isnan value
print( 'isnan check:', df_x_train.isna().any() )
# build a list to clean both train and test
list2clean = [ df_x_test, df_x_train]
for mydata in list2clean:
    #print(mydata)
    # turn male/female, embarked to a number
    mydata.loc[mydata.Sex == 'female','Sex' ] = 0
    mydata.loc[mydata.Sex == 'male','Sex' ] = 1
    mydata.loc[mydata.Embarked == 'S','Embarked' ] = 0
    mydata.loc[mydata.Embarked == 'C','Embarked' ] = 1
    mydata.loc[mydata.Embarked == 'Q','Embarked' ] = 2
    mydata.loc[mydata.Survived == 1, 'Survived ' ] = 'Survived'
    mydata.loc[mydata.Survived == 0, 'Survived ' ] = 'Died'
    # replace nan with average
    mydata.loc[df_x_train.Age.isnull(),'Age']  = np.mean( mydata.Age[~mydata.Age.isna()] )
    mydata.loc[mydata.Embarked.isna(),'Embarked'] = mydata.mode().Embarked[0]
    #print(mydata)
df_x_test= list2clean[0]
df_x_train = list2clean[1]
#print(list2clean[0])
print( 'isnan check:', df_x_train.isna().any() )
```

```python
mode_embark = stats.mode( np.array(df_x_test.Embarked.values) )
print( mode_embark )
```

```python
# make some scatter plots
plt.scatter( df_x_train.Age[y_train == 0], df_x_train.Pclass[y_train == 0], c='r' )
plt.scatter( df_x_train.Age[y_train == 1], df_x_train.Pclass[y_train == 1], c='b' )
plt.show()
```
